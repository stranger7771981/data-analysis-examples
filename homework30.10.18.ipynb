```python
import pandas as pd
```

```python
df = pd.read_csv('data/minjust2018.csv',
                header=None,
                usecols=[4,12,13,14,16,17,19,23,26,27,28,33])
```

```python
df.rename(columns={4:"name", 
                   27:"director", 
                   28:"activity",
                   33:"founder",
                   12:"region",
                   13:"district",
                   14:"city",
                   16:"street",
                   19:"tel"},
         inplace = True)
df.fillna('', inplace = True)
```

```python
#Найти все компании, которые находятся на вашей улице

df[df['street'].str.contains('Федорова')]
```

```python
#Найти все компании, которые принадлежат вашим тезкам
df[df['founder'].str.contains('Денис')]
```

```python
#Найти все компании, которые находятся на вашей улице и принадлежат вашим тезкам
df[df['street'].str.contains('Федорова')&
df['founder'].str.contains('Денис')]
```

```python
#Найти две компании, принадлежащие вашим друзьям, знакомым, родителям
df[df['founder'].str.contains('Углов')]
   
```

```python
#Найти все компании, которые занимаются строительством и в своем телефоне имеют цифры 432 подряд
df[df['activity'].str.contains('строительство')&
df['tel'].str.contains('432')]
```
